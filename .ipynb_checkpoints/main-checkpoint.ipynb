# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
# create engine to hawaii.sqlite
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()

# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# View all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

```python
# View columns of Measurement class
measdata = session.query(Measurement).first()
measdata.__dict__
```

```python
# View columns of Station class
statdata = session.query(Station).first()
statdata.__dict__
```

# Exploratory Precipitation Analysis

```python
# Find the most recent date in the data set.
#from sqlalchemy import asc, desc
recent = session.query(Measurement.date).order_by(Measurement.date.desc()).first()
last_date = recent[0]
last_date
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results. 
# Starting from the most recent data point in the database. 

# Calculate the date one year from the last date in data set.
year_ago = dt.date(2017,8,23) - dt.timedelta(days=365)

# Perform a query to retrieve the data and precipitation scores
twelvemos = session.query(Measurement.date, Measurement.prcp).\
    filter(Measurement.date > year_ago).\
    filter(Measurement.date <= last_date)

#     #view for correctness of query
# for line in twelvemos:
#     print(line)

# Save the query results as a Pandas DataFrame and set the index to the date column
dates = []
prcp = []
    # store records where precipitation amount is not null to lists
for line in twelvemos:
    if type(line[1]) == float:
        dates.append(line[0])
        prcp.append(line[1])

    # Create the dataframe, set index
df = pd.DataFrame({"Date": dates, "Precipitation Score": prcp})
df = df.set_index("Date")


# Sort the dataframe by date
df = df.sort_index(ascending = True)
    # View dataframe
df
# Use Pandas Plotting with Matplotlib to plot the data
precplot = df.plot(kind = "line", figsize= (20,10), rot = 90)
precplot.set_ylabel("Inches")
precplot.set_title(f'All Station Precipitation Over 1 Year')
plt.tight_layout()
plt.savefig('Images/precipitation.png')
plt.show()
```

```python
# Use Pandas to calculate the summary statistics for the precipitation data
df.describe()
```

# Exploratory Station Analysis

```python
# Design a query to calculate the total number stations in the dataset
session.query(Station.id).count()
```

```python
# Design a query to find the most active stations (i.e. what stations have the most rows?)
# List the stations and the counts in descending order.

# Used this code first to obtain list
station_query = engine.execute(\
    'select Station.id, Measurement.station, count(Measurement.station)\
         from Station\
         join Measurement ON Measurement.station = Station.station\
         group by Measurement.station\
         order by count(Measurement.station) desc'\
              ).fetchall()

# returns top station name to be used below
top_station = station_query[0][1]



## Tried another way using SQLAlchemy language to query
# station_query = session.query(Station.id, Measurement.station, func.count(Measurement.station)).\
#         filter(Measurement.station == Station.station).\
#         group_by(Measurement.station).order_by(func.count(Measurement.station).desc()).\
#         all()


# top_station = station_query[0][1]

station_query

```

```python
# Using the most active station id from the previous query, calculate the lowest, highest, and average temperature.
from sqlalchemy import func

low = session.query(func.min(Measurement.tobs)).filter(Measurement.station == top_station).one()
high = session.query(func.max(Measurement.tobs)).filter(Measurement.station == top_station).one()
avg = session.query(func.avg(Measurement.tobs)).filter(Measurement.station == top_station).one()

print(f'The lowest temperature at station {top_station} is {low[0]}.')
print(f'The highest temperature at station {top_station} is {high[0]}.')
print(f'The average temperature at station {top_station} is {round(avg[0],2)}.')
```

```python
# Using the most active station id
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
twelvemosactive = session.query(Measurement.date, Measurement.tobs).\
    filter(Measurement.date > year_ago).\
    filter(Measurement.date <= last_date).\
    filter(Measurement.station == top_station)

dates = []
tobs = []
for line in twelvemosactive:
    if type(line[1]) == float:
        dates.append(line[0])
        tobs.append(line[1])

df = pd.DataFrame({"Date": dates, "Temperatures": tobs})
df = df.set_index("Date")


# Sort the dataframe by date
df = df.sort_index(ascending = True)
df
# Use Pandas Plotting with Matplotlib to plot the data
precplot = df.plot(kind = "hist", figsize= (10,5))
precplot.set_xlabel("Temperature")
precplot.set_title(f'Station {top_station} Temps Over 1 Year')
plt.tight_layout()
plt.savefig('Images/active.png')
plt.show()
```
